## Neuron

인간의 뇌에서 신호를 전달하는 역할을 수행하게 됨.

입력 신호들이 들어왔을 때, 이 신호들의 총 크기가 어떤 값, threshold을 넘게 되면 다음 뉴런으로 전파됨.
그렇지 않다면 전파되지 않음. 이를 본따 만든 것이 인공 신경망.

$\phi ( \sum x \cdot w + b )$ ($\phi$: activation function, e.g. sigmoid)

- **Linear Perceptron**

- AND, OR 문제를 해결하기 위해 만들어 짐.

- AND, OR 풀 수 있으니 더 복잡한 것도 풀 수 있겠구나 하고 사람들이 기대를 가지면서 이 뉴럴 네트워크를 바라보게 되었는데, 1969년 Marvin Lee Minsky가 한 개 레이어로 구성된 퍼셉트론으로는 XOR 문제를 해결할 수 없다는 것을 증명했고, MLP를 쌓아야지만 이 XOR를 풀 수 있다. 하지만, 이 weight들을, 각 층이 여러 개의 weight들을 학습할 방법이 존재하지 않는다는 것을 증명함.

- 그래서 엄청난 기대를 받던 인공 신경망이 암흑기에 빠지게 됨. 지금이야 멀티 레이어를 학습할 수 있는, back propagation이라는 방법이 개발이 되어서 잘 사용하고 있지만, 당시에는 학습할 수 있는 방법이 없어서, 뉴럴 네트워크가 주춤하게 되는 암흑기에 빠지게 됨.

|A|B|XOR|
|:-:|:-:|:-:|
|0|0|0|
|0|1|1|
|1|0|1|

## xor

```python
# Lab 9 XOR
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# for reproducibility
torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)

# nn layers
linear = torch.nn.Linear(2, 1, bias=True)
sigmoid = torch.nn.Sigmoid()

# model
model = torch.nn.Sequential(linear, sigmoid)

# define cost/loss & optimizer
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1)

for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)
    # cost/loss function
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    if step % 100 == 0:
        print(step, cost.item())
```

```python
# Accuracy computation
# True if hypothesis>0.5 else False
with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == Y).float().mean()
    print('\nHypothesis: ', hypothesis.detach().cpu().numpy(), '\nCorrect: ', predicted.detach().cpu().numpy(), '\nAccuracy: ', accuracy.item())
```

- 200번째 epoch쯤 이후부터는 loss 값이 전혀 줄어들지 않고, 일정한 값을 가짐을 확인할 수 있고, 이 말은 즉, 학습이 제대로 되지 않음을 말해 주는 것.

- 그리고, 퍼셉트론이 0.5만 output으로 내고 있음을 확인할 수 있고, 즉 모든 값을 0.5로 예측하고 있고, 정확도가 0.5에서 더 향상이 되지 않는다!
